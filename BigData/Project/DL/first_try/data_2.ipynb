```python
# 트레인 / 테스트 Arrest 비율 맞추기 용
# 체포 확률 20프로 데이터셋으로 ... 맞추기 
testDF = VehicleDF  # 기존 데이터 안건들이기 목적 복사

# sample(n=데이터양)
TrainFalseDF = VehicleDF[VehicleDF['Arrest'] == False].sample(n=100000, random_state=42)
TrainTrueDF = VehicleDF[VehicleDF['Arrest'] == True].sample(n=25000, random_state=42)


# 트레인 데이터로 뽑지 않은 데이터로 다시 테스트 데이터 뽑기
testDF.drop(TrainFalseDF.index, inplace=True)
testDF.drop(TrainTrueDF.index, inplace=True)

TestFalseDF = testDF[testDF['Arrest'] == False].sample(n=8000, random_state=42)
TestTrueDF = testDF[testDF['Arrest'] == True].sample(n=2485, random_state=42)
```

```python
TrainDF = pd.concat([TrainFalseDF, TrainTrueDF], axis=0)
TestDF = pd.concat([TestFalseDF, TestTrueDF], axis=0)
```

```python
TrainDF.to_csv('TrainDF.csv')
TestDF.to_csv('TestDF.csv')
```
