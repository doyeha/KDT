```python
import os
import pandas as pd

# 수정할 폴더 경로 설정
folder_path = r'D:\data\anootations\replace/'  # 폴더 경로를 실제 경로로 변경

# 폴더 내 모든 txt 파일을 반복
for filename in os.listdir(folder_path):
    if filename.endswith('.txt'):
        file_path = os.path.join(folder_path, filename)

        # 파일을 읽어와서 데이터프레임으로 변환
        df = pd.read_csv(file_path, sep=' ', header=None)

        # 첫 번째 열에서 1과 2를 스위칭
        df[0] = df[0].replace({1: 2, 2: 1})

        # 수정된 데이터프레임을 동일한 파일에 다시 저장
        df.to_csv(file_path, sep=' ', header=False, index=False)

print("모든 파일이 성공적으로 수정되었습니다.")
```
