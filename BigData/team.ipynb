```python
import itertools
import random

# 4명씩 가능한 조합을 찾기
possible_combinations = list(itertools.combinations(members, 4))

# 중복 횟수 및 함께 해본 사람들 추적 함수
def calculate_group_overlap(group, df):
    overlap = 0
    overlap_members = []
    for i in range(len(group)):
        for j in range(i + 1, len(group)):
            current_overlap = df.loc[group[i], group[j]]
            overlap += current_overlap
            if current_overlap > 0:
                overlap_members.append((group[i], group[j], current_overlap))
    return overlap, overlap_members

# 중복 횟수가 4 이하인 조합 찾기
valid_groups = []

for group in possible_combinations:
    overlap, overlap_members = calculate_group_overlap(group, df)
    if overlap <= 4:  # 중복 점수를 4로 제한
        valid_groups.append((group, overlap, overlap_members))

# 8개의 조를 랜덤으로 선택
random.shuffle(valid_groups)  # 조합을 섞음
final_groups = []

# 중복 경험이 2회인 사람들을 추적
pairs_to_avoid = set()

# 중복 횟수 2 이상인 조합을 추적
for group in valid_groups:
    overlap_members = group[2]
    for pair in overlap_members:
        if pair[2] == 2:  # 중복 횟수가 2인 경우
            pairs_to_avoid.add((pair[0], pair[1]))

# 조를 구성
while len(final_groups) < 8 and valid_groups:
    group_data = valid_groups.pop()
    group = group_data[0]

    # 조원 중 중복 경험이 2회인 경우 제외
    if all((member not in itertools.chain(*pairs_to_avoid) for member in group)):
        if all(member not in itertools.chain(*[g[0] for g in final_groups]) for member in group):  # 중복 멤버 없을 때만 추가
            final_groups.append(group_data)

# 결과 출력
if len(final_groups) < 8:
    print(f"충분한 조합을 찾지 못했습니다. 가능한 조합 수: {len(final_groups)}")
else:
    for i, (group, overlap, overlap_members) in enumerate(final_groups):
        print(f"{i+1}조: {group} (중복 횟수: {overlap})")
        if overlap_members:
            print("  함께 해본 사람들:")
            for pair in overlap_members:
                print(f"    {pair[0]} - {pair[1]}: {pair[2]}번")
        else:
            print("  함께 해본 경험 없음")
```
