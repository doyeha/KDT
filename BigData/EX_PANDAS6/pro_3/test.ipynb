#### 연도별로 범죄 유형(살인, 강도, 절도, 성폭력 등)의 발생 빈도 변화를 분석
특정 범죄 유형이 다른 범죄 유형보다 더 많은 추세를 보이는지 분석
지역별 범죄 발생률 비교

법률 여부 / 범죄 카테고리 / 범죄명

```python
crime_year = pd.read_excel('범죄발생_월_20_22.xlsx')
crime_year.columns = ['법률 여부', '범죄 카테고리', '범죄명',2020,2021,2022]    # 칼럼명 변경
crime_year # 확인용 호출

# crime_year.do
```

```python
sogyeDF = crime_year[(crime_year['범죄명']=='소계') & (crime_year['법률 여부']=='형법범')]

sogDF = sogyeDF.copy()
# sogDF[2020].count()
# sogDF['증감']=[

# ]
# sogDF['증감'] = "+" if sogDF[2020]-sogDF[2021] > 0 else "-", "+" if sogDF[2021]-sogDF[2022] > 0 else "-"

test01 = []
for i in range(0,sogDF[2020].count()):
    if sogDF.iloc[i][2020]-sogDF.iloc[i][2021] > 0:
        if sogDF.iloc[i][2021]-sogDF.iloc[i][2022] > 0:
            test01.append("↘↘")
        elif sogDF.iloc[i][2021]-sogDF.iloc[i][2022] < 0:
            test01.append("↘↗")
        else:
            test01.append("↘→")

    elif sogDF.iloc[i][2020]-sogDF.iloc[i][2021] < 0:
        if sogDF.iloc[i][2021]-sogDF.iloc[i][2022] > 0:
            test01.append("↘↗")
        elif sogDF.iloc[i][2021]-sogDF.iloc[i][2022] < 0:
            test01.append("↗↗")
        else:
            sogDF.iloc[i]['증감'] = "↘→"


sogDF['증감'] = test01
# sogDF.iloc[0][2020] - sogDF[0][2021]

#     sogDF[i]['증감'] = ("+" if sogDF[i][2020]-sogDF[i][2021] > 0 else "-", "+" if sogDF[i][2021]-sogDF[i][2022] > 0 else "-")
sogDF
```
