#### [ Matplotlib 활용 데이터 시각화]
 - X축과 Y축 다양한 처리

```python
import pandas as pd
from matplotlib import pyplot as plt
```

```python
# [2] 그래프 그리기
y_data = [1,3,5,7,9]
x_data = ['2024-01-01','2024-02-01','2024-03-01','2024-04-01','2024-05-01']
plt.plot(x_data, y_data, marker = 'o')
plt.ylim(bottom = 0, top = 15)
plt.xlim(right = '2023-12-01', left = '2024-06-01')
plt.show()
```

```python
# [2] 그래프 그리기
y_data = [1,3,5,7,9]
x_data = ['2024-01-01','2024-02-01','2024-03-01','2024-04-01','2024-05-01']
plt.plot(x_data, y_data, marker = 'o')
plt.xticks(rotation = 45)
plt.show()
```

```python
# [2] 그래프 그리기  - 그래프 출력 창 즉, figure를 자동 맞춤
y_data = [1,3,5,7,9]
x_data = ['2024-01-01','2024-02-01','2024-03-01','2024-04-01','2024-05-01']

plt.figure(figsize=[10,5])
plt.plot(x_data, y_data, marker = 'o')
plt.xticks(rotation = 45)
plt.show()
```

```python
# [2] 그래프 그리기  - 그래프 출력 창 즉, figure를 자동 맞춤
y_data = [1,3,5,7,9]
x_data = ['2024-01-01','2024-02-01','2024-03-01','2024-04-01','2024-05-01']

# plt.figure(figsize=[10,5])
plt.plot(x_data, y_data, marker = 'o')
plt.tight_layout()  # 모든 내용 출력될 수 있도록 자동 맞춤 설정 
plt.show()
```

```python
# [3] 그래프 그리기  - 그래프 출력 창 즉, figure를 자동 맞춤
y_data = [1,3,5,7,9]
x_data = ['2024-01-01','2024-02-01','2024-03-01','2024-04-01','2024-05-01']

# plt.figure(figsize=[10,5])
plt.plot(x_data, y_data, marker = 'o')
plt.xlim(left = 0, right = 0)
plt.ylim(bottom = 0, top = 15)

# plt.tight_layout()  # 모든 내용 출력될 수 있도록 자동 맞춤 설정 

font_config = {'fontsize' : 22, 'fontweight':'bold'}
plt.xticks([0,2,4,6], ['A','B','C','D'], **font_config)

plt.show()
```
