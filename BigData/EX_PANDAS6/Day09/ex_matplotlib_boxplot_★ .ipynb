```python
# [2] 데이터 준비
F_DATA = r'C:\Git\KDT\BigData\EX_PANDAS6\datasets\auto_mpg.csv'
```

```python
dataDF = pd.read_csv(F_DATA)
dataDF.head(1)
```

```python
# 컬럼별 타입 확인
dataDF.dtypes
```

```python
# [4] 특정 컬럼의 데이터 비율
# - 자동차 생산국가에 대한 비율
dataDF['origin'].astype('category')
```

```python
# - 자동차 생산국가에 대한 비율
round(dataDF['origin'].value_counts()/dataDF['origin'].count(),2)
```

```python
originSR = dataDF['origin'].value_counts()
x = originSR.index.to_list()
h = originSR.values.tolist()
x,h
```

```python
# 자동차 생산국가에 대한dataDF['origin'].unique 비율 시각화 - bar그래프
origin_Bar = plt.bar(x, h, color=['red','yellow','blue'], edgecolor = ['black'])

# 막대에 라벨을 설정
plt.bar_label(origin_Bar, h, label_type='center')    # 바 위에 수치 올려두기

#x축 눈금 설정
plt.xticks([1,2,3],['US','EU','JP'], rotation=25)       # [1][2] 자리에서  1은 눈금 자리, 2는 눈금에 들어갈 값
plt.suptitle('자동차 생산국가')
plt.show()
```

```python
# - 자동차 생산국가에 대한 비율 시각화 - pie 그래프
plt.pie(h, labels=['US','EU','JP'], autopct='%.1f') # autopct가 퍼센트 알아서 계산해주니까 형식 지정만 해주면 된다.
plt.show()
```

```python
# - 자동차 생산국가에 대한 비율 시각화 - pie 그래프
plt.pie(h, labels=['US','EU','JP'], autopct='%.1f',
        counterclock=False, explode=(0.1,0.1,0)) # autopct가 퍼센트 알아서 계산해주니까 형식 지정만 해주면 된다.
plt.show()
```

```python
# - 자동차 생산국가에 대한 비율 시각화 - pie 그래프
plt.pie(h, labels=['US','EU','JP'], autopct='%.1f',
        counterclock=False, wedgeprops={'linewidth':3 ,'width': 0.4, 'edgecolor':'w'}
        # width가 원 두께, 1이 우리가 아는 기본이고 수가 작아질수록 표의 원이 얇아짐
        ) # autopct가 퍼센트 알아서 계산해주니까 형식 지정만 해주면 된다.
plt.show()
```
